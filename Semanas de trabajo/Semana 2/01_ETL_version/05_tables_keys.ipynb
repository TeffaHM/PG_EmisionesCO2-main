```python
import numpy as np
import pandas as pd
```

```python
country = pd.read_csv('../Modelo_ER/country.csv')
emission = pd.read_csv('../Modelo_ER/emission.csv')
plant = pd.read_csv('../Modelo_ER/plant.csv')
population_country_electricity = pd.read_csv('../Modelo_ER/population_country_electricity.csv')
source_country_electricity = pd.read_csv('../Modelo_ER/source_country_electricity.csv')
source_energy = pd.read_csv('../Modelo_ER/source_energy.csv')
region = pd.read_csv('../Modelo_ER/region.csv')
emission_region = pd.read_csv('../Modelo_ER/emission_region.csv')
source_region_electricity = pd.read_csv('../Modelo_ER/source_region_electricity.csv')
```

### Acoplando clave en emission (idCountry, idSource)

```python
emission.insert(0, 'idCountry', '')
emission.insert(3, 'idSource', '')
```

```python
# idCountry
emission = emission.merge(country, on = ['country'])
emission.head()
```

```python
emission['idCountry_x'] = emission['idCountry_y']
emission = emission.rename(columns = {'idCountry_x':'idCountry'})
emission.drop(['idCountry_y', 'iso_code', 'country'], axis = 1, inplace=True)
```

```python
# idSource
emission = emission.merge(source_energy, on = ['source'])
emission.head()
```

```python
emission['idSource_x'] = emission['idSource_y']
emission = emission.rename(columns = {'idSource_x':'idSource'})
emission.drop(['idSource_y', 'source'], axis = 1, inplace=True)
```

```python
emission.to_csv('../Modelo_ER_2/emission.csv', index=False)
```

### Acoplando clave en plant (idCountry)

```python
plant.insert(0, 'idCountry', '')
plant.insert(1, 'idPrimaryFuel', '')
```

```python
# idCountry
plant = plant.merge(country, on = ['country'])
plant.head()
```

```python
plant['idCountry_x'] = plant['idCountry_y']
plant = plant.rename(columns = {'idCountry_x':'idCountry'})
plant.drop(['idCountry_y', 'iso_code', 'country'], axis = 1, inplace=True)
```

```python
# idPrimaryFuel
plant = plant.merge(source_energy, left_on = ['primaryfuel'], right_on = ['source'])
plant.head()
```

```python
plant['idPrimaryFuel'] = plant['idSource']
plant.drop(['idSource', 'source', 'primaryfuel'], axis = 1, inplace=True)
```

```python
plant
```

```python
plant.sort_values(['idCountry','namePlant']).to_csv('../Modelo_ER_2/plant.csv', index=False)
```
